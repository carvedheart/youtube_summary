```python
from pytubefix import Playlist, Channel, YouTube
import pandas as pd
```

```python
# url = input("Nhap url: ")
url = "https://www.youtube.com/@2y0x"
# url = "https://www.youtube.com/playlist?list=PLWKjhJtqVAblStefaz_YOVpDWqcRScc2s"
```

```python
def channel_or_playlist(url):

    if "youtube.com" not in url and "youtu.be" not in url:
        return "Kh√¥ng ph·∫£i URL YouTube"
    
    # Chu·∫©n h√≥a URL b·∫±ng c√°ch th√™m https:// n·∫øu thi·∫øu
    if not url.startswith(("http://", "https://")):
        url = "https://" + url

    from urllib.parse import urlparse, parse_qs
    
    parsed_url = urlparse(url)
    query_params = parse_qs(parsed_url.query)

    # Ki·ªÉm tra URL playlist (c√≥ th·ªÉ c√≥ d·∫°ng list=... ho·∫∑c l√† ƒë∆∞·ªùng d·∫´n playlist)
    if "list" in query_params:
        return "Playlist"
    if "/playlist" in parsed_url.path:
        return "Playlist"
    
    # Ki·ªÉm tra URL channel (c√≥ th·ªÉ c√≥ d·∫°ng /channel/, /c/, /user/, ho·∫∑c @username)
    if "/channel/" in parsed_url.path:
        return "Channel"
    if "/c/" in parsed_url.path:
        return "Channel"
    if "/user/" in parsed_url.path:
        return "Channel"
    if "/@" in parsed_url.path:
        return "Channel"
    
    # N·∫øu kh√¥ng ph·∫£i playlist hay channel
    return "Video th√¥ng th∆∞·ªùng ho·∫∑c kh√¥ng x√°c ƒë·ªãnh"
    
```

```python
def url_Listing(url):
    if channel_or_playlist(url) == "Playlist":
        playlist = Playlist(url)
        data = []
        for video in playlist.videos:  # S·ª≠ d·ª•ng .videos thay v√¨ .video_urls
            try:
                data.append({
                    "Title": video.title,
                    "URL": video.watch_url,  # S·ª≠ d·ª•ng watch_url ƒë·ªÉ l·∫•y URL
                    "Duration": video.length,
                    "Author": video.author
                })
            except Exception as e:
                print(f"L·ªói khi x·ª≠ l√Ω video {video.watch_url}: {str(e)}")
                continue
        return data

    elif channel_or_playlist(url) == "Channel":
        channel = Channel(url)
        data = []
        for video in channel.videos:  # S·ª≠ d·ª•ng .videos thay v√¨ .video_urls
            try:
                data.append({
                    "Title": video.title,
                    "URL": video.watch_url,  # S·ª≠ d·ª•ng watch_url ƒë·ªÉ l·∫•y URL
                    "Duration": video.length,
                    "Author": video.author
                })
            except Exception as e:
                print(f"L·ªói khi x·ª≠ l√Ω video {video.watch_url}: {str(e)}")
                continue
        return data

    else:
        return None
```

```python
data = url_Listing(url)
data
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('video_list')
```
